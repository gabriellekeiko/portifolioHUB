## 4) Na coluna `Código da Emenda`, substitua todos os valores `"Sem informação"` por `"Não disponível"`. Mostre a contagem de valores únicos da coluna antes e depois da substituição.

```python
unicos_antes = df['Código da Emenda'].nunique()
df['Código da Emenda'] = df['Código da Emenda'].replace('Sem informação', 'Não disponível')
unicos_depois = df['Código da Emenda'].nunique()

print(f"Valores únicos antes: {unicos_antes}")
print(f"Valores únicos depois: {unicos_depois}")
```

## 5) Na coluna `Nome Função`, transforme todos os valores para que fiquem em letras maiúsculas. Confirme a alteração mostrando os valores únicos da coluna.

```python
df['Nome Função'] = df['Nome Função'].str.upper()

print("Valores únicos na coluna 'Nome Função':")
print(df['Nome Função'].unique())
```

## 6) Verifique quantas linhas duplicadas existem no dataset. Em seguida, remova as duplicatas e mostre quantas linhas sobraram.

```python
total_duplicadas = df.duplicated().sum()
print(f"Linhas duplicadas encontradas: {total_duplicadas}")

df = df.drop_duplicates()

linhas_restantes = len(df)
print(f"Linhas restantes após a remoção: {linhas_restantes}")
```

## 7) Remova duplicatas considerando apenas as colunas `Código da Emenda` e `Município`. Quantas linhas sobraram? O resultado é diferente de quando removemos duplicatas do DataFrame inteiro?

```python
df_subset = df.drop_duplicates(subset=['Código da Emenda', 'Município'])
linhas_subset = len(df_subset)

df_inteiro = df.drop_duplicates()
linhas_inteiro = len(df_inteiro)

print(f"Linhas após remover por Código/Município: {linhas_subset}")
print(f"Linhas após remover duplicatas totais: {linhas_inteiro}")
```

O resultado é DIFERENTE. Existem linhas que compartilham o mesmo Código e Município, mas têm valores diferentes em outras colunas.

## 8) Identifique quais colunas possuem valores nulos e quantos nulos cada uma tem. Em seguida, responda: qual é a coluna que possui maior número de valores nulos e quantos são?

```python
nulos_por_coluna = df.isnull().sum()
colunas_com_nulos = nulos_por_coluna[nulos_por_coluna > 0]

print("Colunas com valores nulos:")
print(colunas_com_nulos)

coluna_max_nulos = nulos_por_coluna.idxmax()
quantidade_max_nulos = nulos_por_coluna.max()

print(f"\nColuna com mais nulos: {coluna_max_nulos}")
print(f"Quantidade: {quantidade_max_nulos}")
```

## 9) Preencha os valores nulos da coluna `codigo do autor` com **-1** e os valores nulos da coluna `nome do autor da emenda` com "desconhecido" e confirme que não restaram nulos nessa coluna.

```python
df.loc[:, 'Código do Autor da Emenda'] = df['Código do Autor da Emenda'].fillna(-1)
df.loc[:, 'Nome do Autor da Emenda'] = df['Nome do Autor da Emenda'].fillna('desconhecido')

nulos_codigo = df['Código do Autor da Emenda'].isnull().sum()
nulos_nome = df['Nome do Autor da Emenda'].isnull().sum()

print(f"Nulos em 'Código do Autor da Emenda': {nulos_codigo}")
print(f"Nulos em 'Nome do Autor da Emenda': {nulos_nome}")
```

## 10) Verifique o tipo da coluna `Ano da Emenda`. Converta-a para **string** (`str`). Mostre o tipo antes e depois da conversão.

```python
tipo_antes = df['Ano da Emenda'].dtype
print(f"Tipo antes: {tipo_antes}")

df['Ano da Emenda'] = df['Ano da Emenda'].astype(str)

tipo_depois = df['Ano da Emenda'].dtype
print(f"Tipo depois: {tipo_depois}")
```

## 11) Converta a coluna `Valor Empenhado` para o tipo **float** (caso ainda não seja). Mostre o tipo após a conversão.

```python
df = df.copy()
df['Valor Empenhado'] = df['Valor Empenhado'].astype(float)

print(f"Tipo da coluna 'Valor Empenhado' após a conversão: {df['Valor Empenhado'].dtype}")
```

## 12) Calcule o **total** de `Valor Pago` por **Região**. Quais são as duas regiões que mais receberam recursos?

```python
df['Valor Pago'] = df['Valor Pago'].astype(float)
total_por_regiao = df.groupby('Região')['Valor Pago'].sum().sort_values(ascending=False)

print("Total de Valor Pago por Região:")
print(total_por_regiao)
```

As duas regiões que mais receberam recursos são Nordeste e Sudeste

## 14) Crie uma nova coluna chamada `Saldo a Pagar`. Ela deve ser o resultado da subtração entre a coluna `Valor Empenhado` e a coluna `Valor Pago`. Mostre as 5 primeiras linhas apenas com essas três colunas.
